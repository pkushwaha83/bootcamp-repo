```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data_df = pd.merge(study_results, mouse_metadata,
                                 how='outer', on='Mouse ID')

# Display the data table for preview
combined_data_df.head()

#GroupMiceData = combined_data_df.groupby(["Mouse ID", "Timepoint"])

#MiceAvgs = GroupMiceData.mean()
#MiceAvgs.head()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#remove g989 data from both data sets
clean_data_frame = combined_data_df.loc[combined_data_df["Mouse ID"] != "g989", :]
clean_data_frame.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

GrugGroup = combined_data_df.groupby(['Drug Regimen'])
#Mean Tumor Volume
MeanTV = GrugGroup['Tumor Volume (mm3)'].mean()
#Median Tumor Volume
MedianTV = GrugGroup['Tumor Volume (mm3)'].median()
#Variance Tumor Volume
VarTV = GrugGroup['Tumor Volume (mm3)'].var()
#Standard deviation Tumor Volume
StdTV = GrugGroup['Tumor Volume (mm3)'].std()
#SEM deviation
StdErrTV = GrugGroup['Tumor Volume (mm3)'].sem()
          
SumTV_df = pd.DataFrame({"Mean Tumor Volume":MeanTV,"Median Tumor Volume":MedianTV,"Tumor Volume Variance":VarTV,"Tumor Volume St. Dev":StdTV,"Tumor Volume St. Err":StdErrTV})
SumTV_df = SumTV_df[["Mean Tumor Volume","Median Tumor Volume","Tumor Volume Variance","Tumor Volume St. Dev","Tumor Volume St. Err"]]
SumTV_df.head()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function

mean_sem_tv = combined_data_df.groupby(['Drug Regimen']).agg({"Tumor Volume (mm3)" :["mean", "median" ,"var","std", "sem"]})
mean_sem_tv.head()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
clean_data_frame["Mouse ID"].value_counts()

#locate the g989 rows in the drug data
l509Drug = clean_data_frame.loc[clean_data_frame["Mouse ID"] == "l509", :]
l509Drug
groupedDrugData = l509Drug.groupby(["Drug Regimen", "Timepoint"])

#find time point versus tumor volume for each drug at each timepoint
drugAvgs = groupedDrugData.mean()

#list timepoints
timepoints = np.arange(0, 50, 5)

#list tumor vol avg by drug and timepoint
tumorVolAvg = drugAvgs["Tumor Volume (mm3)"]

#separate tumor vol avg list by drug
capo_tumorVolAvg = tumorVolAvg.loc["Capomulin", :]

#plot each drug's tumor vol avg over time
capo_tumorVolAvgPlot = plt.plot(timepoints, capo_tumorVolAvg, marker = "D", label = "Capomulin", alpha = .75)


#formatting and titles
plt.hlines(45, 0, 45, alpha=0.25)
plt.title("Capomulin Treatment of Mouse l509")
plt.xlabel("Timepoint Days")
plt.ylabel("Tumor Volume (mm^3)")
plt.legend(loc="best")
plt.tight_layout()
plt.show()


```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
Capomulin_df = clean_data_frame.loc[clean_data_frame["Drug Regimen"] == "Capomulin", :]

mouse_weight = Capomulin_df.groupby(Capomulin_df["Mouse ID"])["Weight (g)"].mean()
tumor_volume = Capomulin_df.groupby(Capomulin_df["Mouse ID"])["Tumor Volume (mm3)"].mean()

#plot each drug's tumor vol avg over time
capo_mouseweightAvgPlot = plt.scatter(mouse_weight,tumor_volume, marker = "o", label = "Capomulin",  alpha = .75)


#formatting and titles
#plt.title("Capomulin Treatment of Mouse l509")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm^3)")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

Capomulin_df = clean_data_frame.loc[clean_data_frame["Drug Regimen"] == "Capomulin", :]
# Pull values for x and y values
Mouseweight = Capomulin_df.groupby(Capomulin_df["Mouse ID"])["Weight (g)"].mean()
Tumorvolume = Capomulin_df.groupby(Capomulin_df["Mouse ID"])["Tumor Volume (mm3)"].mean()

# Perform a linear regression on year versus violent crime rate
slope, int, r, p, std_err = st.linregress(Mouseweight, Tumorvolume)
      
# Create equation of line to calculate predicted violent crime rate
fit = slope * mouse_weight + int

# Plot the linear model on top of scatter plot 
plt.scatter(Mouseweight,Tumorvolume)
plt.xlabel("Weight of Mouse")
plt.ylabel("Tumor Volume")
plt.plot(Mouseweight,fit,"--")
plt.xticks(Mouseweight, rotation=90)
plt.show()

# Caculate correlation coefficient
corr = round(st.pearsonr(Mouseweight,Tumorvolume)[0],2)
print(f'The correlation between weight and tumor value is {corr}')
```
